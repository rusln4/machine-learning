```python
# штука для открытия pdf
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return "\n".join(text)
```

```python
def parse_habr_pdf(text):
    lines = [line.strip() for line in text.split('\n') if line.strip()]
    
    data = {
        'NameCompany': '',
        'Raiting': 0,
        'DataPublish': '',
        'Activity': '',
        'TextArticle': text
    }
    
    # Поиск рейтинга и названия компании
    for i, line in enumerate(lines):
        if re.match(r'^\d+\.?\d*$', line) and i+1 < len(lines) and "Рейтинг" in lines[i+1]:
            try:
                data['Raiting'] = float(line)
            except ValueError:
                data['Raiting'] = 0
            
            if i+2 < len(lines):
                data['NameCompany'] = lines[i+2]
            if i+3 < len(lines):
                data['Activity'] = lines[i+3]
            break
    
    # Поиск даты публикации
    date_patterns = [
        # для относительного времени
        (r'(?:\w+\s)?(\d+\s(?:час[ао]в|дн[ея]|минут)\sназад)', lambda m: m.group(1)),
        
        # для полных дат
        (r'(\d{1,2}\s(?:января|февраля|марта|апреля|мая|июня|июля|августа|'
        r'сентября|октября|ноября|декабря)\s\d{4})', lambda m: m.group(1)),
        
        # сокращенные названия месяцев (дек, янв и т.д.)
        (r'(\d{1,2}\s(?:янв|фев|мар|апр|мая|июн|июл|авг|сен|окт|ноя|дек)[а-я]*\s\d{4})'
        r'(?:\sв\s\d{2}:\d{2})?', lambda m: m.group(1)),
        
        # в формате DD.MM.YYYY
        (r'(\d{2}\.\d{2}\.\d{4})(?:,\s\d{2}:\d{2})?', lambda m: m.group(1)),
        
        # в формате YYYY-MM-DD
        (r'(\d{4}-\d{2}-\d{2})', lambda m: m.group(1))
    ]
    
    for line in lines:
        for pattern, handler in date_patterns:
            match = re.search(pattern, line)
            if match:
                data['DataPublish'] = handler(match)
                break
        if data['DataPublish']:
            break
    return data
```

```python
import re
# Обработка всех PDF файлов
results = []
failed_files = []

for pdf_path in all_pdf:
    try:
        # Извлекаем текст (теперь это строка, а не список)
        text = extract_text_from_pdf(pdf_path)
        # print(*text)
        # print('\n\n\n')
        
        parsed_data = parse_habr_pdf(text)
        if parsed_data['NameCompany'] == '':
            continue
        results.append(parsed_data)
        
        print(f"Успешно обработан: {pdf_path.split('\\')[-1]}")
    except Exception as e:
        failed_files.append((pdf_path, str(e)))
        print(f"ОШИБКА !!!!!!!!!! {pdf_path.split('\\')[-1]}: {str(e)}")
```
